```python
train=df[df["Date"] <"20150101"]
test=df[df["Date"]>"20141231"]
```

```python
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
```

```python
list1=[str(x) for x in range(25)]
data.columns=list1
```

```python
data.head()
```

```python
list2=[int(x) for x in range(3975)]
```

```python
for i in list1:
    data[i]=data[i].str.lower()
```

```python
" ".join(x for x in data.iloc[0,:25])
```

```python
headlines=[]
```

```python
for i in range(0,len(data.index)):
    headlines.append(" ".join(str(x) for x in data.iloc[i,:25]))
```

```python
len(headlines)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector=CountVectorizer(ngram_range=(2,2))
```

```python
traindataset=countvector.fit_transform(headlines)
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion="entropy")
randomclassifier.fit(traindataset,train.Label)
```

```python
test_headlines=[]
for i in range(0,len(test.index)):
    test_headlines.append(" ".join(str(x) for x in test.iloc[i,2:27]))
```

```python
test.info()
```

```python
len(test_headlines)
```

```python
testdataset=countvector.transform(test_headlines)
```

```python
testdataset
```

```python
prediction=randomclassifier.predict(testdataset)
```

```python
randomclassifier.score(testdataset,test.Label)
```
